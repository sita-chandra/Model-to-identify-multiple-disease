# **⭐ Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# visualize pairplot
import seaborn as sns
sns.heatmap(disease.corr())
```

```python
# column names
disease.columns
```

```python
# define y
y=disease.prognosis
y.shape
```

```python
# define X
pX=disease.drop(['fluid_overload','prognosis'], axis=1)
X = pX.loc[:, ~pX.columns.duplicated()]
X.shape
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# verify shape
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=200)
```

```python
# train model
model.fit(X_train,y_train)
```

```python
# predict with model
y_pred=model.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# model accuracy
accuracy_score(y_test,y_pred)
```

```python
# model confusion matrix
confusion_matrix(y_test,y_pred)
```

```python
#confusion matrix visualization
import seaborn as sns
sns.heatmap(confusion_matrix(y_test,y_pred))
```

```python
# model classification report
print(classification_report(y_test,y_pred))
```
